```python
import pandas as pd
import plotly.graph_objects as go
```

```python
source = pd.read_csv('/content/test.csv')
target = pd.read_csv('/content/v.0.1.0.1331.csv')
target['class'] = target['class'].apply(lambda x: ' ' + x)
```

```python
merged_df = source.merge(target, on='id', suffixes=('_source', '_target'))
labels = list(set(merged_df['class_source']).union(set(merged_df['class_target'])))
label_index = {label: idx for idx, label in enumerate(labels)}


sources = merged_df['class_source'].map(label_index).tolist()
targets = merged_df['class_target'].map(label_index).tolist()
values = [1] * len(merged_df)

# Create the Sankey diagram
fig = go.Figure(go.Sankey(
    node=dict(
        pad=15,
        thickness=20,
        line=dict(color="black", width=0.2),
        label=labels,
    ),
    link=dict(
        source=sources,
        target=targets,
        value=values,
    )
))

fig.update_layout(title_text="Class Changes from Source.csv to Target.csv", font_size=10)
fig.show()
```
